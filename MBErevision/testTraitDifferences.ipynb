```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
```

```python
df = pd.read_csv("finData.tab", sep="\t", header=0)
df
```

```python
targets = df[df.phenotypic_group == "target"]
targets
```

```python
outgroups = df[df.phenotypic_group == "outgroup"]
outgroups
```

```python
mannwhitneyu(targets.num_pelvic_fins, outgroups.num_pelvic_fins, alternative='two-sided')
```

```python
mannwhitneyu(targets.num_caudal_rays, outgroups.num_caudal_rays, alternative='two-sided')
```
